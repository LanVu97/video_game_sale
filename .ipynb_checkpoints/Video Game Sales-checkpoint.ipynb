# Introduction
Fields include

Rank - Ranking of overall sales

Name - The games name

Platform - Platform of the games release (i.e. PC,PS4, etc.)

Year - Year of the game's release

Genre - Genre of the game

Publisher - Publisher of the game

NA_Sales - Sales in North America (in millions)

EU_Sales - Sales in Europe (in millions)

JP_Sales - Sales in Japan (in millions)

Other_Sales - Sales in the rest of the world (in millions)

Global_Sales - Total worldwide sales.

Link: https://www.kaggle.com/gregorut/videogamesales


# Import data

```python
import pandas as pd
raw_df = pd.read_csv('vgsales.csv')
raw_df
```

```python
raw_df.columns
```

# Data Preparation & Cleaning


```python
df = raw_df.copy()
```

```python
df.info()
```

```python
# convert datatype of year to int64
df['Year'] = df['Year'].convert_dtypes()
```

```python
df.dtypes
```

```python
df.shape
```

```python
# Drop the column 'Rank'
df = df.drop(columns = 'Rank')
```

```python
df.duplicated().unique()
```

Dataset don't have duplicated value.

```python
# list missing value
df.isnull().sum()
```

```python
# rows 'Global_Sales' don't match sum of sale of 'NA_Sales','EU_Sales','JP_Sales',and 'Other_Sales'
df[df['NA_Sales'] + df['EU_Sales'] + df['JP_Sales'] + df['Other_Sales'] != df['Global_Sales']]
```

It have 2 issues:
- missing data in year and Publisher column
- 'Global_Sales' don't equal to sum of sale of 'NA_Sales','EU_Sales','JP_Sales',and 'Other_Sales'

```python
# drop row contain missing value in year and publisher column
df.drop(df[df.Year.isnull()].index, inplace=True)
df.drop(df[df.Publisher.isnull()].index, inplace=True)
```

```python
# test again
df.isnull().sum()
```

```python
# Repair value for 'Global_Sales' column
df['Global_Sales'] = df['NA_Sales'] + df['EU_Sales'] + df['JP_Sales'] + df['Other_Sales']
```

```python
# test again
df[df['NA_Sales'] + df['EU_Sales'] + df['JP_Sales'] + df['Other_Sales'] != df['Global_Sales']]
```

```python
df.describe()
```

### Exploratory Analysis and Visualization
Let's begin by importing matplotlib.pyplot and seaborn.

```python
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

sns.set_style('darkgrid')
matplotlib.rcParams['font.size'] = 14
matplotlib.rcParams['figure.figsize'] = (9, 5)
matplotlib.rcParams['figure.facecolor'] = '#00000000'
```

#### Year
Let's look at total video game release each year.

```python
# List total game release each year
year = df.value_counts('Year').reset_index()
year.columns = ['year', 'counts_game']
year = year.sort_values('year')
```

```python
top_year = year.sort_values('counts_game', ascending= False).head(5)
top_year
```

```python
# We can visualize this information using a line chart
plt.plot('year', 'counts_game' , data = year)
# sns.countplot('Year', data = df)
plt.title('Total Game Release Each Year')
plt.show()
```

It appears that from 2007 to 2011, there are a lot game released. After 2011, numbers of the video games are declining and almost equal to 0 from 2017 to 2020.

#### Region Area
Let calculate sum of sale by each region area

```python
#  total sales each country
region_sale = df[['NA_Sales','EU_Sales','JP_Sales','Other_Sales']].sum()
```

```python
region_sale
```

```python
# We can visualize this information using a bar chart
region = ['North America', 'Eroupe', 'Japan', 'Other']
plt.title('Sales Comparison Between Region ')
plt.bar(region, region_sale, color = '#3274A1')
plt.show()
```

We can see that the North America have the largest sale followed by the EU and JP.

### Q: In the top Video Game Sales in each Region, what type of Genre does the video game belong to? (List by each region )

```python
top_sale_na = df.sort_values('NA_Sales', ascending= False).head(10)
top_sale_na = top_sale_na['Genre'].value_counts().reset_index()
top_sale_na.columns = ['platform', 'counts_na']
top_sale_na
```

```python
top_sale_eu = df.sort_values('EU_Sales', ascending= False).head(10)
top_sale_eu = top_sale_eu['Genre'].value_counts().reset_index()
top_sale_eu.columns = ['platform', 'counts_eu']
top_sale_eu
```

```python
top_sale_jp = df.sort_values('JP_Sales', ascending= False).head(10)
top_sale_jp = top_sale_jp['Genre'].value_counts().reset_index()
top_sale_jp.columns = ['platform', 'counts_jp']
top_sale_jp
```

```python
region_df = top_sale_na.merge(top_sale_jp, on= 'platform', how='outer').merge(top_sale_eu, on= 'platform', how='outer')
region_df = region_df.fillna(0)
region_df['total'] = region_df['counts_na'] + region_df['counts_jp'] + region_df['counts_eu']
region_df = region_df.convert_dtypes()
region_df
```

```python
plt.figure(figsize=(20,10), dpi= 60)
plt.bar('platform', 'counts_na', label = 'North America', color ='#01A2D9', data = region_df)
plt.bar('platform', 'counts_jp', bottom = 'counts_na', label = 'Japan', color = '#014D64', data = region_df)
plt.bar('platform', 'counts_eu', bottom = region_df['counts_jp'] + region_df['counts_na'], label= 'Eroupe',color ='#6794A7', data = region_df)

plt.title('Top Video Game Sales In Each Region by Genre')
plt.legend()
plt.show()
```

Let's see top Video Game Sales each region:
- In Japan market, the genre of video games is less and the most genre is Role-playing.
- In NA market and EU market, the video game genre is more various.
